# Performance on other datasets

```python
import sys
import clip
import os
import os.path as osp

import numpy as np
import torch.nn as nn
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
import torch
from torch.utils.data import DataLoader, Dataset
from torchvision import datasets, transforms
from torchvision.io import read_image
from PIL import Image
del Report
from artemis.emotions import ARTEMIS_EMOTIONS
import EmotionPredictor.data_tools as dt
from EmotionPredictor.evaluation import Report
```

```python
%config InlineBackend.figure_format ='retina'
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
clip_RN50x16, preprocess = clip.load("RN50x16")
clip_RN50x16 = clip_RN50x16.eval()
```

```python

class SLP(nn.Module):
    def __init__(self,input_size = 512, output_size = 9):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(input_size, output_size)
        )
        
    def forward(self, x):
        fp = self.layers(x.float())
        return fp
    
data_path = "../data/DIRTI/"

MODEL_PATH = "../neural_checkpoints/clip_training/RN50x16/model_9" 
slp = SLP(input_size=768)
slp.load_state_dict(torch.load(MODEL_PATH))
slp.eval()
slp.to(device)
```

```python
# We implement our own data loader, the torch version was only reading a subset of the pictures
def get_iterator(path, preprocess) :
    def iterator():
        for image in os.listdir(path):
            if image[-4:] != ".jpg":
                continue
            img_path = osp.join(path,image)
            yield preprocess(Image.open(img_path)).unsqueeze(0)
    return iterator()
```

```python
results = []
with torch.no_grad():
    for img in get_iterator(data_path, preprocess):
        img_features = clip_RN50x16.encode_image(img.to(device))
        emotions = slp(img_features)
        results.append(emotions)
```

```python
results = torch.cat(results)
```

```python
labels = []
for name in os.listdir(data_path):
    if name.split()[-1] == "neutral.jpg" : labels.append(0) 
    elif name[-4:] == ".jpg": labels.append(1)
```

```python
preds = (results.argmax(1).cpu().numpy() == 5).astype(int)
labels = np.array(labels)

conf= pd.DataFrame(metrics.confusion_matrix(labels, preds), index = ["other", "disgust"])
conf.columns= conf.index
#save_latek("../results/tables/dirti_small_conf.tex", conf.to_latex())
conf
```

```python
preds = results.argmax(1).cpu().numpy().astype(int)
labels = np.array(labels)

conf= pd.DataFrame(metrics.confusion_matrix(labels*5, preds, labels=range(0,9)), index = ARTEMIS_EMOTIONS)#["other", "disgust"])
conf.columns= conf.index
#ct.save_latek("../results/tables/dirti_big_conf.tex", conf.to_latex())
```

```python
dis = results.softmax(1)[labels.astype(bool)][:,5].cpu().numpy()
no_dis = results.softmax(1)[~labels.astype(bool)][:,5].cpu().numpy()
```

```python
plt.boxplot([no_dis, dis], )
plt.xticks([1,2], ["neutral", "disgusting"]);
plt.ylabel("disgust score")
plt.grid(axis = "y", ls = ":", alpha = 0.5)
plt.savefig("../results/figures/dirti_box.pdf")
```

```python
dis_scores = results.cpu()[:,5]
dis_scores[[labels.astype(bool)]] = -np.inf 
dis_scores.argmax()
direc = list(os.listdir(data_path))
direc.remove('.ipynb_checkpoints')
print("Neutral picture provoking the most disgust:")
Image.open(osp.join(data_path, direc[dis_scores.argmax()]))
```
